## 2. Hardware
- Revolution in Customized Hardware for DL gave a significant boost to DL's success.
- Till 2010, CPUs wre being used for every DL project
like for matrix multiplication , but then researchers realized that the same can be done through parallel processing, which is  a much faster way.
- NVIDIA started buliding GPU specifically for this,
    - 1993: Founded and started making 3D graphics for gaming and multimedia
    - 1999: started making GPUs
    - 2006: CUDA (language for GPUs)for parallel processing
    - 2012; AI breakthrough Alexnet(CNN architecture, significantly advanced image recognition)
    - and so on...
- GPUs reduces training time upto 10 to 20 Times

- **ASIC: Application Specific Integrated Circuits**
      - Faster, optimized for specific tasks
      - Ultra low power
      - fixed function , not reprogrammable (unlike FGPAs)
      - Not Flexible, Fixed functionality cannot be chnaged after fabrication
      - Higher upfront cost (then FGPAs for specific tasks) --(Higher initial cost, lower per unit cost in mass production)
  - Company
        - TSMC

              - processors

  - Types of ASICs:

        1. TPUs (Tensor Processing Unit):
          Google's AI-specific chip for accelerating deep learning workloads (e.g., TensorFlow)
        2. Edge TPUs:
          A smaller, power-efficient TPU designed for AI inference on edge devices (e.g., IoT, mobile)
        3. NPUs (Neural processing Unit):
          AI-optimized accelerator found in mobile devices, used for on-device machine learning
